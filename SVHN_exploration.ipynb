# Setup

```python
import glob
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

# Read Data

```python
path_train = 'data/SVHN/train/train'
path_test = 'data/SVHN/test/test'

images_train = glob.glob(os.path.join(path_train, '*.png'))
images_test = glob.glob(os.path.join(path_test, '*.png'))
print(f'Nr. of train images: {len(images_train)}')
print(f'Nr. of test images: {len(images_test)}')
```

# Example plots

```python
# training set
fig, axes = plt.subplots(2, 10, figsize=(15,3))

for i, ax in enumerate(axes.flatten()):
    im = Image.open(images_train[i])
    ax.imshow(im)
    ax.axis('off')
```

```python
im.size
```

```python
# test set
fig, axes = plt.subplots(2, 10, figsize=(15,3))

for i, ax in enumerate(axes.flatten()):
    im = Image.open(images_test[i])
    ax.imshow(im)
    ax.axis('off')
```

# Notes

* Images have different sizes

```python
# training set
fig, axes = plt.subplots(2, 10, figsize=(15,3))

for i, ax in enumerate(axes.flatten()):
    im = Image.open(images_train[i])
    im = im.resize((64, 64))
    ax.imshow(im)
    ax.axis('off')
```

# Plot GAN Results

```python
path = 'data/SVHN'
fake_images = np.load(os.path.join(path, 'fake_images_201e_wgan.npy'))
real_images = np.load(os.path.join(path, 'real_images_201e_wgan.npy'))
```

```python
print(f'fake images shape: {fake_images.shape}')
print(f'real images shape: {real_images.shape}')
```

```python
fig, axes = plt.subplots(2, 1, figsize=(15,6))
for i, ax in enumerate(axes.flatten()):
    im = real_images[i]
    ax.imshow(im)
    ax.axis('off')
```

```python
fig, axes = plt.subplots(10, 2, figsize=(15,32))
for i, ax in enumerate(axes.flatten()):
    im = fake_images[i]
    ax.imshow(im)
    ax.set_title(f'epoch {i*10+1}')
    ax.axis('off')
```
